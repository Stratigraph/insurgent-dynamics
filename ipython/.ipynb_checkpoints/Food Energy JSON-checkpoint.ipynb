```python
groups = pd.read_csv(masterpath+'groups.csv',';')
groups.head()
```

```python
itemcount=dict() #initialize dictionary of counts for food items for every food group
itemvalue=dict() #initialize dictionary of values for food items for every food group
links=pd.DataFrame()
nodes=pd.DataFrame()
for key in list(set(groups['Group'].values)):
        itemcount[key]=0
        itemvalue[key]=0
count = 0
for i in data[(data.Element=='Production')&(data.Year==1961)&(data.Country=='Brazil')&(~data.Item.str.contains('Total'))].values:
    fooditem=i[3] #choose column in which food item names are stored, might need to change it
    foodvalue=i[9]
    if (fooditem in groups['FCL Title'].values):
        #print fooditem, groups['Group'].values[groups[groups['FCL Title']==fooditem].index[0]]
        nodes.set_value(count,"name",fooditem)
        links.set_value(count,"source",fooditem)
        links.set_value(count,"target",groups['Group'].values[groups[groups['FCL Title']==fooditem].index[0]])
        links.set_value(count,"value",foodvalue)
        count+=1
        itemcount[groups['Group'].values[groups[groups['FCL Title']==fooditem].index[0]]]+=1
        itemvalue[groups['Group'].values[groups[groups['FCL Title']==fooditem].index[0]]]+=foodvalue
    else:
        if (',' in fooditem):
            if (fooditem[0:fooditem.find(',')] in groups['FCL Title'].values):
                #print fooditem[0:fooditem.find(',')], groups['Group'].values[groups[groups['FCL Title']==fooditem[0:fooditem.find(',')]].index[0]]
                count+=1
                itemcount[groups['Group'].values[groups[groups['FCL Title']==fooditem[0:fooditem.find(',')]].index[0]]]+=1
                itemvalue[groups['Group'].values[groups[groups['FCL Title']==fooditem[0:fooditem.find(',')]].index[0]]]+=foodvalue
        else:
            if ('(' in fooditem):
                if (fooditem[0:fooditem.find('(')] in groups['FCL Title'].values):
                    #print fooditem[0:fooditem.find('(')], groups['Group'].values[groups[groups['FCL Title']==fooditem[0:fooditem.find('(')]].index[0]]
                    count+=1 
                    itemcount[groups['Group'].values[groups[groups['FCL Title']==fooditem[0:fooditem.find('(')]].index[0]]]+=1
                    itemvalue[groups['Group'].values[groups[groups['FCL Title']==fooditem[0:fooditem.find('(')]].index[0]]]+=foodvalue
#print count, data[(data.Element=='Production')&(data.Year==1961)&(data.Country=='Brazil')&(~data.Item.str.contains('Total'))].Item.values.size
#print itemcount
#print itemvalue

links
```

```python
year = 1961
country = 'Brazil'
itemcount=dict() #initialize dictionary of counts for food items for every food group
itemvalue=dict() #initialize dictionary of values for food items for every food group
jsondict={'nodes':[],'links':[]} #initialize dictionary for later JSON export for D3 Sankey visualization

for key in list(set(groups['Group'].values)):
        itemcount[key]=0
        itemvalue[key]=0

count = 0

for i in data[(data.Element=='Production')&(data.Year==year)&(data.Country==country)&(~data.Item.str.contains('Total'))].values:
    fooditem=i[3]  #choose column in which food item names are stored, might need to change it
    foodvalue=i[9] #choose column in which food item values are stored, might need to change it
    nodes_innerdict={}
    links_innerdict={}
    if (fooditem in groups['FCL Title'].values):
        foodname = fooditem
        foodgroup = groups['Group'].values[groups[groups['FCL Title']==fooditem].index[0]]
        count+=1
        itemcount[groups['Group'].values[groups[groups['FCL Title']==fooditem].index[0]]]+=1
        itemvalue[groups['Group'].values[groups[groups['FCL Title']==fooditem].index[0]]]+=foodvalue
    else:
        if (',' in fooditem):
            if (fooditem[0:fooditem.find(',')] in groups['FCL Title'].values):
                foodname = fooditem[0:fooditem.find(',')]
                foodgroup = groups['Group'].values[groups[groups['FCL Title']==fooditem[0:fooditem.find(',')]].index[0]]
                count+=1
                itemcount[groups['Group'].values[groups[groups['FCL Title']==fooditem[0:fooditem.find(',')]].index[0]]]+=1
                itemvalue[groups['Group'].values[groups[groups['FCL Title']==fooditem[0:fooditem.find(',')]].index[0]]]+=foodvalue
        else:
            if ('(' in fooditem):
                if (fooditem[0:fooditem.find('(')] in groups['FCL Title'].values):
                    foodname = fooditem[0:fooditem.find('(')]
                    foodgroup = groups['Group'].values[groups[groups['FCL Title']==fooditem[0:fooditem.find('(')]].index[0]]
                    count+=1 
                    itemcount[groups['Group'].values[groups[groups['FCL Title']==fooditem[0:fooditem.find('(')]].index[0]]]+=1
                    itemvalue[groups['Group'].values[groups[groups['FCL Title']==fooditem[0:fooditem.find('(')]].index[0]]]+=foodvalue

    nodes_innerdict["name"]=foodname
    links_innerdict["source"]=foodname
    links_innerdict["target"]=foodgroup
    links_innerdict["value"]=foodvalue
        
    jsondict["nodes"].append(nodes_innerdict)
    jsondict["links"].append(links_innerdict)

#print count, data[(data.Element=='Production')&(data.Year==1961)&(data.Country=='Brazil')&(~data.Item.str.contains('Total'))].Item.values.size
#print itemcount
#print itemvalue
file('../json/'+country+repr(year)+'.json','w').write(repr(jsondict))
```
