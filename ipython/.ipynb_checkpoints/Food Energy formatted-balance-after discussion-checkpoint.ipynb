Select indicator and import data from the CSV from within the zip file. This can take anywhere from a few minutes to several hours, depending on your computer's performance. If you have access to paralell computing capabilites please turn them on at the *Clusters* tab of the main Ipython window. 

```python
ind =6
if localpath:
    r = masterpath+indicators[ind]+'.zip'
    z = zipfile.ZipFile(r)
else:
    r = requests.get(masterpath+indicators[ind]+'.zip')
    z = zipfile.ZipFile(StringIO.StringIO(r.content))
data = pd.read_csv(z.open(indicators[ind]+'.csv'))
data.head()
```

```python
elements=['Food','Feed','Seed','Processing','Waste','Other Util']
#for country in set(data.Country.values):
country='World + (Total)'    
if True:
    #for year in set(data[data.Country==country].Year.values):
    year = 1961
    if True:    
        count = 0
        groupdict={}
        groupsum={}
        groupvalue={}
        
        jsondict={"nodes":[],"links":[]} #initialize dictionary for later JSON export for D3 Sankey visualization
        for element in elements:
            jsondict["nodes"].append({"name":element})
            if element not in groupdict:
                groupdict[element]=count
                count += 1
        for element in elements:
            for i in data[(data.Element==element)&(data.Year==year)&(data.Country==country)&(data.Item.str.contains('Total'))].values:
                foodgroup = 'Vegetal Products'
                if (i[2]>2940): #choose column in which food group codes are stored, might need to change it
                    foodgroup = 'Animal Products'
                if (i[2]>2959):
                    foodgroup = 'Fish & Seafood'
                fooditem=i[3]  #choose column in which food item names are stored, might need to change it
                foodvalue=i[9] #choose column in which food item values are stored, might need to change it
                  
                if fooditem not in groupdict: 
                    groupdict[fooditem]=count
                    jsondict["nodes"].append({"name":fooditem[0:fooditem.find('+')-1]})
                    count += 1  
                
                links_innerdict={}    
                links_innerdict["source"]=groupdict[fooditem]
                links_innerdict["target"]=groupdict[element]
                links_innerdict["value"]=foodvalue
                links_innerdict["supply"]=foodvalue*2
                
                if (foodgroup != 'Vegetal Products'):
                    links_innerdict={} 
                    links_innerdict["source"]=groupdict['Feed']
                    links_innerdict["target"]=groupdict[fooditem]
                    links_innerdict["value"]=foodvalue
                    links_innerdict["supply"]=foodvalue*2  
               
                if (~np.isnan(foodvalue)&(foodvalue!=0)):
                    if foodgroup in groupsum: 
                        groupsum[foodgroup]+=foodvalue
                    else:
                        groupsum[foodgroup]=foodvalue
                    if element in groupsum: 
                        groupsum[element]+=foodvalue
                    else:
                        groupsum[element]=foodvalue
                    jsondict["links"].append(links_innerdict)
                    
                '''   
                if (foodgroup != 'Vegetal Products'):
                    links_innerdict={} 
                    links_innerdict["source"]=groupdict['Feed']
                    links_innerdict["target"]=groupdict[fooditem]
                    links_innerdict["value"]=foodvalue
                    links_innerdict["supply"]=foodvalue*2
                    if (~np.isnan(foodvalue)&(foodvalue!=0)):
                        jsondict["links"].append(links_innerdict)
                '''
             
        links_innerdict={} 
        links_innerdict["source"]=groupdict['Processing']
        links_innerdict["target"]=groupdict['Food']
        links_innerdict["value"]=groupsum['Processing']
        links_innerdict["supply"]=groupsum['Processing']*2
        if (~np.isnan(groupsum['Processing'])&(groupsum['Processing']!=0)):
            jsondict["links"].append(links_innerdict)
             
        file('../json/'+country+repr(year)+'.json','w').write(json.dumps(jsondict)) 

print jsondict
```
