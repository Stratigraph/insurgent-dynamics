Select indicator and import data from the CSV from within the zip file. This can take anywhere from a few minutes to several hours, depending on your computer's performance. If you have access to paralell computing capabilites please turn them on at the *Clusters* tab of the main Ipython window. 

```python
ind =6
if localpath:
    r = masterpath+indicators[ind]+'.zip'
    z = zipfile.ZipFile(r)
else:
    r = requests.get(masterpath+indicators[ind]+'.zip')
    z = zipfile.ZipFile(StringIO.StringIO(r.content))
data = pd.read_csv(z.open(indicators[ind]+'.csv'))
data.head()
```

```python
#for country in set(data.Country.values):
country='World + (Total)'    
if True:
    #for year in set(data[data.Country==country].Year.values):
    year = 1961
    if True:    
        count = 0
        groupdict={}
        groupsum={}
        jsondict={"nodes":[],"links":[]} #initialize dictionary for later JSON export for D3 Sankey visualization
        for i in data[(data.Element=='Production')&(data.Year==year)&(data.Country==country)&(~data.Item.str.contains('Total'))].values:
            fooditem=i[3]  #choose column in which food item names are stored, might need to change it
            foodvalue=i[9] #choose column in which food item values are stored, might need to change it
            foodgroup=i[2] #choose column in which food group codes are stored, might need to change it
            links_innerdict={}
            found = False
            if (fooditem in groups['FCL Title'].values):
                found = True
                foodname = fooditem
            else:
                if (',' in fooditem):
                    if (fooditem[0:fooditem.find(',')] in groups['FCL Title'].values):
                        found = True
                        foodname = fooditem[0:fooditem.find(',')]
                    else:
                        if ('(' in fooditem):
                            if (fooditem[0:fooditem.find('(')] in groups['FCL Title'].values):
                                found = True
                                foodname = fooditem[0:fooditem.find('(')]
    
            if found:
                foodgroup = groups['Group'].values[groups[groups['FCL Title']==foodname].index[0]]
                foodgroup = foodgroup[foodgroup.find('-')+2:len(foodgroup)].capitalize()
                
                if foodgroup in groupsum: 
                    groupsum[foodgroup]+=foodvalue
                else:
                    groupsum[foodgroup]=foodvalue
        
                if fooditem not in groupdict: 
                    groupdict[fooditem]=count
                    jsondict["nodes"].append({"name":fooditem})
                    count += 1
                if foodgroup not in groupdict:  
                    groupdict[foodgroup]=count
                    jsondict["nodes"].append({"name":foodgroup})
                    count += 1
                                                
                links_innerdict["source"]=groupdict[fooditem]
                links_innerdict["target"]=groupdict[foodgroup]
                links_innerdict["value"]=foodvalue
                links_innerdict["supply"]=foodvalue*2
                jsondict["links"].append(links_innerdict)
             
        links_innerdict={}
        jsondict["nodes"].append({"name":"Agricultural products"})
       
        for i in groupsum.keys():
            links_innerdict={}
            links_innerdict["source"]=groupdict[i]
            links_innerdict["target"]=count
            links_innerdict["value"]=groupsum[i]
            links_innerdict["supply"]=groupsum[i]*2
            jsondict["links"].append(links_innerdict)
        
        file('../json/'+country+repr(year)+'d.json','w').write(json.dumps(jsondict))
        
        count = 0;
        jd={"nodes":[],"links":[]}
        jd["nodes"].append({"name":"Agricultural products"})
        for i in groupsum.keys():
            count += 1
            jd["nodes"].append({"name":i})
            links_innerdict={}
            links_innerdict["source"]=count
            links_innerdict["target"]=0
            links_innerdict["value"]=groupsum[i]
            links_innerdict["supply"]=groupsum[i]*2
            jd["links"].append(links_innerdict)
        
        file('../json/'+country+repr(year)+'.json','w').write(json.dumps(jd))
        
        print jd
```
