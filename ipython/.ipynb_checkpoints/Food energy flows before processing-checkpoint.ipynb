```python
def get_data(ind):
    if localpath:
        r = masterpath+indicators[ind]+'.zip'
        z = zipfile.ZipFile(r)
    else:
        r = requests.get(masterpath+indicators[ind]+'.zip')
        z = zipfile.ZipFile(StringIO.StringIO(r.content))
    return pd.read_csv(z.open(indicators[ind]+'.csv'))
balance = get_data(6)
prod = get_data(10)
supply1 = get_data(7)
supply2 = get_data(8)
print set(balance.Element)
print set(prod.Element)
print set(supply.Element)
```
