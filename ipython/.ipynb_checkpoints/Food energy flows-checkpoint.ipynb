```python
def get_data(ind):
    if localpath:
        r = masterpath+indicators[ind]+'.zip'
        z = zipfile.ZipFile(r)
    else:
        r = requests.get(masterpath+indicators[ind]+'.zip')
        z = zipfile.ZipFile(StringIO.StringIO(r.content))
    return pd.read_csv(z.open(indicators[ind]+'.csv'))

balance = get_data(6)
energy = get_data(3)
fertilizers1 = get_data(15)
fertilizers2 = get_data(16)
population = get_data(9)
print set(balance.Element)
print set(energy.Element)
print set(fertilizers1.Element)
print set(fertilizers2.Element)
print set(population.Element)
```

```python
def add_to_json(data,i,source,target,value,prod,supply):
    global groupdict
    global count
    if (~np.isnan(value)&(value!=0)):
        if source not in groupdict[i]: 
            groupdict[i][source]=count[i]
            data[i]["nodes"].append({"name":source})
            count[i] += 1 
        if target not in groupdict[i]: 
            groupdict[i][target]=count[i]
            data[i]["nodes"].append({"name":target})
            count[i] += 1 
        links_innerdict={} 
        links_innerdict["source"]=groupdict[i][source]
        links_innerdict["target"]=groupdict[i][target]
        links_innerdict["value"]=value
        links_innerdict["prod"]=prod
        links_innerdict["supply"]=supply
        data[i]["links"].append(links_innerdict)

#elements1=['Feed','Food','Processing','Other Util','Seed','Waste','Stock Variation','Import Quantity','Export Quantity']
elements1=['Feed','Food','Processing','Other Util','Seed','Waste']
elements2=['Food supply','Food supply quantity (kg/capita']
fuels=['Gas-diesel oils', 'Natural gas (including LNG)', 'Residual fuel oil', 'Liquefied petroleum gas (LPG)']
fert1=['Phosphate fertilizers', 'Potash fertilizers', 'Nitrogenous fertilizers']
fert2=['Potash Fertilizers (K20 total nutrients)', 'Phosphate Fertilizers (P205 total nutrients)', 'Nitrogen Fertilizers (N total nutrients)']

#for country in set(data.Country.values):
for country in ['World + (Total)']:#,'United States of America', 'Canada']:    
   #for year in set(balance[balance.Country==country].Year.values):
   for year in range(1981,2009):
        
        ###############################################################################################
        ######                                   FOOD PART                                       ######
        ###############################################################################################
        
        #initialize food output variable for sum collection and processing for EROEI
        foodoutput=0
        
        # associate food items to their respective groups
        n = 4          #number of grouping variations
        supergroup=[]  #group to which food item entry belongs to. defined below, custom groups possible, along the FAO official ones 
        parent={}      #food item parent(source) node, where applicable (secondary food items), i.e. Feed for Animal Products or Milk for Cheese
        for i in range (n):
            supergroup.append({})
        
        for i in balance[(balance.Element=="Food")&(balance.Year==year)&(balance.Country==country)\
                          &(((2511<=balance.ItemCode)&(balance.ItemCode<=2782))\
                          |(balance.ItemCode==2805)\
                          |(balance.ItemCode==2899)\
                          |(balance.ItemCode==2912)\
                          |(balance.ItemCode==2945)\
                          |(balance.ItemCode==2948)\
                          |(balance.ItemCode==2949)\
                          |(balance.ItemCode==2961))].values:
            fooditem=i[3]         #choose column in which food item names are stored, might need to change it
            if "+" in fooditem:   #simplify food names for the sake of visualization
                fooditem=fooditem[0:fooditem.find("+")-1]
            if "(" in fooditem:
                fooditem=fooditem[0:fooditem.find("(")-1]
            if "and products" in fooditem:
                fooditem=fooditem[0:fooditem.find("and products")-1]
            index = i[2]         #choose column in which food group codes are stored, might need to change it
            parent[fooditem]='None'
            for i in range (n):  #assign default supergroup 'Others' and no parent to all food items
                supergroup[i][fooditem]='Others'
            if ((2511<=index<=2659) | (index==2745) | (index==2805) | (index==2899) | (index==2912)):
                if ((2511<=index<=2520) | (index==2805)):
                    supergroup[1][fooditem] = 'Cereals - Excluding Beer'
                    supergroup[0][fooditem] = 'Cereals'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'None'
                elif (2531<=index<=2535):
                    supergroup[1][fooditem] = 'Starchy Roots'
                    supergroup[0][fooditem] = 'Starchy Roots'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'None'
                elif (2536<=index<=2537):
                    supergroup[1][fooditem] = 'Sugar Crops'
                    supergroup[0][fooditem] = 'Sugar Crops'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'None'
                elif ((2541<=index<=2543)|(index==2745)):
                    supergroup[1][fooditem] = 'Sugar & Sweeteners'
                    supergroup[0][fooditem] = 'Sugar'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'Processing'
                elif (2546<=index<=2549):
                    supergroup[1][fooditem] = 'Pulses'
                    supergroup[0][fooditem] = 'Fruits & Vegetables'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'None'
                elif (2555<=index<=2570):
                    supergroup[1][fooditem] = 'Oilcrops'
                    supergroup[0][fooditem] = 'Oilcrops'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'None'
                elif (2571<=index<=2586):
                    supergroup[1][fooditem] = 'Vegetable Oils'
                    supergroup[0][fooditem] = 'Vegetable Oils'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'Processing'
                elif (2601<=index<=2605):
                    supergroup[1][fooditem] = 'Vegetables'
                    supergroup[0][fooditem] = 'Fruits & Vegetables'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'None'
                elif (2611<=index<=2625):
                    supergroup[1][fooditem] = 'Fruits - Excluding Wine'
                    supergroup[0][fooditem] = 'Fruits & Vegetables'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'None'
                elif (2630<=index<=2635):
                    supergroup[1][fooditem] = 'Stimulants'
                    supergroup[0][fooditem] = 'Others'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'None'
                elif (2640<=index<=2645):
                    supergroup[1][fooditem] = 'Spices'
                    supergroup[0][fooditem] = 'Others'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'None'
                elif (2655<=index<=2659):
                    supergroup[1][fooditem] = 'Alcoholic Beverages'
                    supergroup[0][fooditem] = 'Other Processed'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'Processing'
                elif (index==2899):
                    supergroup[1][fooditem] = 'Miscellaneous'
                    supergroup[0][fooditem] = 'Others'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'None'
                elif (index==2912):
                    supergroup[1][fooditem] = 'Treenuts'
                    supergroup[0][fooditem] = 'Fruits & Vegetables'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'None'
                       
            elif ((2731<=index<=2782) | (index==2945)| (index==2948) | (index==2949) | (index==2961)):
                if (2731<=index<=2736):
                    supergroup[1][fooditem] = 'Meat'
                    supergroup[0][fooditem] = 'Animal Products'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'Feed'
                elif (2737<=index<=2743):
                    supergroup[1][fooditem] = 'Animal Fats'  
                    supergroup[0][fooditem] = 'Other Processed' 
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'Processing'
                elif (2761<=index<=2768):
                    supergroup[1][fooditem] = 'Fish, Seafood'
                    supergroup[0][fooditem] = 'Aquatic Products'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'Feed'
                elif ((2769<=index<=2775)|(index==2961)):
                    supergroup[1][fooditem] = 'Aquatic Products, Other'
                    supergroup[0][fooditem] = 'Aquatic Products'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'Feed'
                elif (2781<=index<=2782):
                    supergroup[1][fooditem] = 'Animal Fats'
                    supergroup[0][fooditem] = 'Other Processed'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'Processing'
                elif (index==2945):
                    supergroup[1][fooditem] = 'Offals'
                    supergroup[0][fooditem] = 'Animal Products'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'Feed'
                elif (index==2948):
                    supergroup[1][fooditem] = 'Milk - Excluding Butter'
                    supergroup[0][fooditem] = 'Animal Products'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'Feed'
                elif (index==2949):
                    supergroup[1][fooditem] = 'Eggs'    
                    supergroup[0][fooditem] = 'Animal Products'
                    parent[fooditem]=parent[supergroup[0][fooditem]]=parent[supergroup[1][fooditem]]= 'Feed'
        
        #extract food item caloric values
        percapitasupply={}
        percapitaenergy={}
        for i in balance[(balance.Element=='Food supply')&(balance.Year==year)&(balance.Country==country)].values:
            foodvalue=i[9]        #choose column in which food item values are stored, might need to change it
            fooditem=i[3]         #choose column in which food item names are stored, might need to change it
            if "+" in fooditem:   #simplify food names for the sake of visualization
                fooditem=fooditem[0:fooditem.find("+")-1]
            if "(" in fooditem:
                fooditem=fooditem[0:fooditem.find("(")-1]
            if "and products" in fooditem:
                fooditem=fooditem[0:fooditem.find("and products")-1]
            percapitaenergy[fooditem]=foodvalue
        
        for i in balance[(balance.Element=='Food supply quantity (kg/capita')&(balance.Year==year)&(balance.Country==country)].values:
            foodvalue=i[9]        #choose column in which food item values are stored, might need to change it
            fooditem=i[3]         #choose column in which food item names are stored, might need to change it
            if "+" in fooditem:   #simplify food names for the sake of visualization
                fooditem=fooditem[0:fooditem.find("+")-1]
            if "(" in fooditem:
                fooditem=fooditem[0:fooditem.find("(")-1]
            if "and products" in fooditem:
                fooditem=fooditem[0:fooditem.find("and products")-1]
            percapitasupply[fooditem]=foodvalue*1000/365.25 #kg/year/capita->g/day/capita
        
        foodcalorie={}
        for i in percapitasupply:
            foodcalorie[i]=100  #set default caloric value for all food items, necessary to handle exceptions
            if i in percapitaenergy:
                if ((percapitasupply[i]==0)|(percapitaenergy[i]==0)): #if either term is null, try to assing the supergroup values, if they exist
                    if ((i in supergroup[1])&(supergroup[1][i] in percapitasupply)&(supergroup[1][i] in percapitaenergy)):
                        if ((percapitasupply[supergroup[1][i]]!=0)&(percapitaenergy[supergroup[1][i]]!=0)):
                            foodcalorie[i]=100*percapitaenergy[supergroup[1][i]]/percapitasupply[supergroup[1][i]]
                else:
                    foodcalorie[i]=100*percapitaenergy[i]/percapitasupply[i]
        
        #extract supply data and calculate energy data
        energydata=[]  #initialize dictionary array for later JSON export for D3 & sankey.js visualization
        supplydata=[]  #initialize dictionary array for later JSON export for D3 & sankey.js visualization
        count = []     #counter for the groupdict below
        groupdict=[{}] #initialize dictionary for mapping food category names to JSON, in the order of addition - necessary format for sankey.js
        parentsupplysum=[]   #dictionary to keep track of item sums in case of aggregations, based on the supergroup variable defined above
        parentenergysum=[]   #dictionary to keep track of item sums in case of aggregations, based on the supergroup variable defined above    
        for i in range (n):
            energydata.append({"nodes":[],"links":[]})
            supplydata.append({"nodes":[],"links":[]})
            count.append(0)
            groupdict.append({})
            supergroup.append({})
            parentsupplysum.append({})
            parentenergysum.append({})
        
        for element in elements1:
            supplysum=[]   #dictionary to keep track of item sums in case of aggregations, based on the supergroup variable defined above
            energysum=[]   #dictionary to keep track of item sums in case of aggregations, based on the supergroup variable defined above
            for i in range (n):
                supplysum.append({})
                energysum.append({})
            
            for i in balance[(balance.Element==element)&(balance.Year==year)&(balance.Country==country)\
                          &(((2511<=balance.ItemCode)&(balance.ItemCode<=2782))\
                          |(balance.ItemCode==2805)\
                          |(balance.ItemCode==2899)\
                          |(balance.ItemCode==2912)\
                          |(balance.ItemCode==2945)\
                          |(balance.ItemCode==2948)\
                          |(balance.ItemCode==2949)\
                          |(balance.ItemCode==2961))].values:
                foodvalue=i[9]            #choose column in which food item values are stored, might need to change it
                if (~np.isnan(foodvalue)&(foodvalue!=0)):
                    fooditem=i[3]         #choose column in which food item names are stored, might need to change it
                    if "+" in fooditem:   #simplify food names for the sake of visualization
                        fooditem=fooditem[0:fooditem.find("+")-1]
                    if "(" in fooditem:
                        fooditem=fooditem[0:fooditem.find("(")-1]
                    if "and products" in fooditem:
                        fooditem=fooditem[0:fooditem.find("and products")-1]
                    
                    energyvalue=foodvalue*foodcalorie[fooditem]*0.00116222222/100
                          #TWh = 1000tons * kcal/100g # 10^9) /100g * 0.00116222222 kwh/kcal *10^-9 TWh/kWh
 
                   #create links for detail level 3    
                    if (element=="Seed"):  #replace seed group with self-loops 
                        add_to_json(supplydata,2,fooditem,fooditem,foodvalue,energyvalue,'[{"l":'+repr(fooditem)+',"v":'+repr(foodvalue)+'}]')   
                        add_to_json(energydata,2,fooditem,fooditem,energyvalue,foodvalue,'[{"l":'+repr(fooditem)+',"v":'+repr(energyvalue)+'}]')   
                    else:
                        add_to_json(supplydata,2,fooditem,element,foodvalue,energyvalue,'[{"l":'+repr(fooditem)+',"v":'+repr(foodvalue)+'}]')
                        add_to_json(energydata,2,fooditem,element,energyvalue,foodvalue,'[{"l":'+repr(fooditem)+',"v":'+repr(energyvalue)+'}]')
                    
                        #sum all links going into food for sepparate representation
                        if (element=="Food"):
                            foodoutput+=energyvalue
                        
                    #create aggregate values for parent links detail level 3
                    if fooditem in parentsupplysum[2]:
                        parentsupplysum[2][fooditem]+=foodvalue
                        parentenergysum[2][fooditem]+=energyvalue
                    else:
                        parentsupplysum[2][fooditem]=foodvalue
                        parentenergysum[2][fooditem]=energyvalue
                    
                    #create aggregate values for detail levels 1-2
                    for i in range(2): #sum food item values over different groups, defined above, currently 1-2
                        #save sum, but also components for the piechart breakdown
                        if supergroup[i][fooditem] in supplysum[i]:
                            supplysum[i][supergroup[i][fooditem]]["data"].append({"l":fooditem, "v":foodvalue})
                            supplysum[i][supergroup[i][fooditem]]["sum"]+=foodvalue
                            energysum[i][supergroup[i][fooditem]]["data"].append({"l":fooditem, "v":energyvalue})
                            energysum[i][supergroup[i][fooditem]]["sum"]+=energyvalue
                        else:
                            supplysum[i][supergroup[i][fooditem]]={"data":[{"l":fooditem, "v":foodvalue}],"sum":foodvalue}
                            energysum[i][supergroup[i][fooditem]]={"data":[{"l":fooditem, "v":energyvalue}],"sum":energyvalue}
            
            #create links for detail levels 1-2    
            for i in range(2):
                for j in supplysum[i]:
                    if (element=="Seed"):
                        add_to_json(supplydata,i,j,j,supplysum[i][j]["sum"],energysum[i][j]["sum"],supplysum[i][j]["data"])
                        add_to_json(energydata,i,j,j,energysum[i][j]["sum"],supplysum[i][j]["sum"],energysum[i][j]["data"])
                    else:
                        add_to_json(supplydata,i,j,element,supplysum[i][j]["sum"],energysum[i][j]["sum"],supplysum[i][j]["data"])
                        add_to_json(energydata,i,j,element,energysum[i][j]["sum"],supplysum[i][j]["sum"],energysum[i][j]["data"])
            
            #create aggregate values for parent links detail levels 1-2
            for i in range(2):
                for j in set(supplysum[i]):
                    if j in parentsupplysum[i]:
                        parentsupplysum[i][j]+=supplysum[i][j]["sum"]
                        parentenergysum[i][j]+=energysum[i][j]["sum"]
                    else:
                        parentsupplysum[i][j]=supplysum[i][j]["sum"]
                        parentenergysum[i][j]=energysum[i][j]["sum"]
                   
        #create parent links for detail levels 1-3
        for i in range(3):    
            for j in parentsupplysum[i]:
                if (parent[j]!="None"):
                    add_to_json(supplydata,i, parent[j],j,parentsupplysum[i][j],parentenergysum[i][j],'[{"l":'+repr( parent[j])+',"v":'+repr(parentsupplysum[i][j])+'}]')
                    add_to_json(energydata,i, parent[j],j,parentenergysum[i][j],parentsupplysum[i][j],'[{"l":'+repr( parent[j])+',"v":'+repr(parentenergysum[i][j])+'}]')
        
        #update nodes name map
        for i in range(n):
            energydata[i]['nodes']=supplydata[i]['nodes']
            
        ###############################################################################################
        ######                                   FUEL PART                                       ######
        ###############################################################################################
        
        #calculate crops share for energy input distribution
        supplyshare=[]
        energyshare=[]
        fuelenergysum=[]
        for i in range (n):
            supplyshare.append({})
            energyshare.append({})
            fuelenergysum.append({})
        
        cropsum=[]
        cropenergy=[]
        for i in range(n):
            cropsum.append(0)
            cropenergy.append(0)
            for j in parentsupplysum[i]:
                if (parent[j]=='None'):
                    cropsum[i] += parentsupplysum[i][j]
                    cropenergy[i] += parentenergysum[i][j]
        
        for i in range(n):
            for j in parentsupplysum[i]:
                if (parent[j]=='None'):
                    supplyshare[i][j] = parentsupplysum[i][j]/cropsum[i]
                    energyshare[i][j] = parentenergysum[i][j]/cropenergy[i]
        
        #distribute electricity input
        for fuel in ['Electricity']:
            for i in energy[(energy.Element=='Consumption in Agriculture')&(energy.Year==year)&(energy.Country==country)\
                            &(energy.Unit=='Terajoule')&(energy.Item==fuel)].values:
                fuelvalue=i[9]            #choose column in which fuel item values are stored, might need to change it
                if (~np.isnan(fuelvalue)&(fuelvalue!=0)):
                    fuelitem=i[3]         #choose column in which fuel item names are stored, might need to change it
                    if "+" in fuelitem:   #simplify fuel names for the sake of visualization
                        fuelitem=fuelitem[0:fuelitem.find("+")-1]
                    if "(" in fooditem:
                        fuelitem=fuelitem[0:fuelitem.find("(")-1]
                    if "and products" in fuelitem:
                        fuelitem=fuelitem[0:fuelitem.find("and products")-1]
                    electricityvalue=fuelvalue*0.000277777778 #1 joule = 0.000277777778 Wh
                    for i in range(n):
                        for j in parentsupplysum[i]:
                            if (parent[j]=='None'):
                                add_to_json(energydata,i, fuel,j,electricityvalue*energyshare[i][j],0,'[{"l":'+repr(fuel)+',"v":'+repr(electricityvalue*energyshare[i][j])+'}]')
         
        #distribute fossil fuel input
        for fuel in fuels:
            for i in energy[(energy.Element=='Consumption in Agriculture')&(energy.Year==year)&(energy.Country==country)\
                            &(energy.Unit=='Terajoule')&(energy.Item==fuel)].values:
                fuelvalue=i[9]            #choose column in which fuel item values are stored, might need to change it
                if (~np.isnan(fuelvalue)&(fuelvalue!=0)):
                    fuelitem=i[3]         #choose column in which fuel item names are stored, might need to change it
                    if "+" in fuelitem:   #simplify fuel names for the sake of visualization
                        fuelitem=fuelitem[0:fuelitem.find("+")-1]
                    if "(" in fooditem:
                        fuelitem=fuelitem[0:fuelitem.find("(")-1]
                    if "and products" in fuelitem:
                        fuelitem=fuelitem[0:fuelitem.find("and products")-1]
                    fuelvalue*=0.000277777778 #1 joule = 0.000277777778 Wh
                    for i in range(n): #sum fuel item values over all groups
                        supergroup[i][fuelitem]="Fossil Fuels"
                        #save sum, but also components for the piechart breakdown
                        if supergroup[i][fuelitem] in fuelenergysum[i]:
                            fuelenergysum[i][supergroup[i][fuelitem]]["data"].append({"l":fuelitem, "v":fuelvalue})
                            fuelenergysum[i][supergroup[i][fuelitem]]["sum"]+=fuelvalue
                        else:
                            fuelenergysum[i][supergroup[i][fuelitem]]={"data":[{"l":fuelitem, "v":fuelvalue}],"sum":fuelvalue}
        f_save=copy.deepcopy(fuelenergysum)
        for i in range(n):
            for j in parentsupplysum[i]:
                fuelenergysum=copy.deepcopy(f_save)
                if (parent[j]=='None'):
                    for k in fuelenergysum[i]:
                        for f in fuelenergysum[i][k]["data"]:
                            f["v"]*=energyshare[i][j]
                        add_to_json(energydata,i,k,j,fuelenergysum[i][k]["sum"]*energyshare[i][j],0,fuelenergysum[i][k]["data"])
         
        
        ###############################################################################################
        ######                                   FERTILIZERS PART                                ######
        ###############################################################################################
        
        fertenergysum=[]
        fertsupplysum=[]
        for i in range (n):
            fertenergysum.append({})
            fertsupplysum.append({})
        
        if (year<2003):
            fertilizers=fertilizers2 #inverted on purpose
            fertnames=fert1
            fertelement='Consumption'
        else:
            fertilizers=fertilizers1
            fertnames=fert2
            fertelement='Consumption in nutrients'
        for fert in fertnames:
            for i in fertilizers[(fertilizers.Element==fertelement)&(fertilizers.Year==year)&(fertilizers.Country==country)\
                            &(fertilizers.Item==fert)].values:
                fertvalue=i[9]            #choose column in which fert item values are stored, might need to change it
                if (~np.isnan(fertvalue)&(fertvalue!=0)):
                    fertitem=i[3]         #choose column in which fert item names are stored, might need to change it
                    #define energy content+packaging and transport (http://www.engr.usask.ca/societies/csae/c9915.pdf page10)
                    if fertitem in ['Phosphate fertilizers', 'Phosphate Fertilizers (P205 total nutrients)']:
                        fertenergycontent=6.82+8.33 #MJ/kg 
                        fertitem='Phosphate'
                    if fertitem in ['Potash fertilizers', 'Potash Fertilizers (K20 total nutrients)']:
                        fertenergycontent=2.88+6.35 #MJ/kg
                        fertitem='Potash'
                    if fertitem in ['Nitrogenous fertilizers', 'Nitrogen Fertilizers (N total nutrients)']:
                        fertenergycontent=59+7.05 #MJ/kg
                        fertitem='Nitrogen'
                    
                    fertvalue=fertvalue/1000 #model units are in 1000 tonnes
                    fertenergy=fertvalue*fertenergycontent         *  0.000277777778   #TWh 
                                          #MJ/kg / TJ/1000tonnes   *  #1 joule = 0.000277777778 Wh
                    for i in range(n): #sum fert item values over all groups
                        supergroup[i][fertitem]="Fertilizers"
                        #save sum, but also components for the piechart breakdown
                        if supergroup[i][fertitem] in fertenergysum[i]:
                            fertenergysum[i][supergroup[i][fertitem]]["data"].append({"l":fertitem, "v":fertenergy})
                            fertenergysum[i][supergroup[i][fertitem]]["sum"]+=fertenergy
                            fertsupplysum[i][supergroup[i][fertitem]]["data"].append({"l":fertitem, "v":fertvalue})
                            fertsupplysum[i][supergroup[i][fertitem]]["sum"]+=fertvalue
                        else:
                            fertenergysum[i][supergroup[i][fertitem]]={"data":[{"l":fertitem, "v":fertenergy}],"sum":fertenergy}
                            fertsupplysum[i][supergroup[i][fertitem]]={"data":[{"l":fertitem, "v":fertvalue}],"sum":fertvalue}
        f_save=copy.deepcopy(fertenergysum)
        for i in range(n):
            for j in parentsupplysum[i]:
                fertenergysum=copy.deepcopy(f_save)
                if (parent[j]=='None'):
                    for k in fertenergysum[i]:
                        for f in fertenergysum[i][k]["data"]:
                            f["v"]*=energyshare[i][j]
                        add_to_json(energydata,i,k,j,fertenergysum[i][k]["sum"]*energyshare[i][j],fertsupplysum[i][k]["sum"]*energyshare[i][j],fertenergysum[i][k]["data"])
           
        ###############################################################################################
        ######                                   LABOR PART                                      ######
        ###############################################################################################
        
        calorietotal=balance[(balance.Element=='Food supply')&(balance.Year==year)&(balance.Country==country)\
                             &(balance.Item=='Grand Total + (Total)')].values[0][9] #second index: choose column in which pop item values are stored, might need to change it
        popvalue=population[(population.Element=='Total economically active population in Agr')&(population.Year==year)\
                            &(population.Country==country)].values[0][9] #second index: choose column in which pop item values are stored, might need to change it
        if (~np.isnan(popvalue)&(popvalue!=0)):
            workyear=5.0/7.0*365.25   #working days spent with farming per year
            
            hourlycalorieburn=204        #calories needed per hour of farming
            hoursperday=12               #hours per day spent farming
            minimumcalorieburn=hourlycalorieburn*hoursperday
            calorieburn=0.9              #percent of calorie intake burned for farming, if below minimum
            
            laborenergy=min(minimumcalorieburn,calorieburn*calorietotal)*workyear*popvalue*1000*0.00116222222/1000000000
                                                                                              # 0.00116222222 kwh/kcal *10^-9 TWh/kWh
            for i in range(n):
                for j in parentsupplysum[i]:
                    if (parent[j]=='None'):
                        add_to_json(energydata,i,"Labor",j,laborenergy*energyshare[i][j],0,'[{"l":'+repr("Labor")+',"v":'+repr(laborenergy*energyshare[i][j])+'}]')
         
        ###############################################################################################
        ######                                   SAVE DATA                                       ######
        ###############################################################################################
        
        for i in range(n):
            file('../json/'+country+repr(year)+'d'+repr(i+1)+'.json','w').write(json.dumps(supplydata[i]))
            file('../json/'+country+repr(year)+'e'+repr(i+1)+'.json','w').write(json.dumps(energydata[i]))
        
        #create global input-output graph    
        fooddata={"nodes":[{"name": "Input"},{"name": "Food"},{"name": "Output"}],"links":[]}
        foodinput=laborenergy+fertenergysum[0]["Fertilizers"]["sum"]+fuelenergysum[0]["Fossil Fuels"]["sum"]+electricityvalue
        
        links_innerdict={} 
        links_innerdict["source"]=0
        links_innerdict["target"]=1
        links_innerdict["value"]=foodinput
        fooddata["links"].append(links_innerdict)
        
        links_innerdict={} 
        links_innerdict["source"]=1
        links_innerdict["target"]=2
        links_innerdict["value"]=foodoutput
        fooddata["links"].append(links_innerdict)
        
        file('../json/'+country+repr(year)+'.json','w').write(json.dumps(fooddata)) 
        
```
